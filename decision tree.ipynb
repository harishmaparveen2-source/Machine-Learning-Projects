```python
#step-1.Import required libraries:

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import seaborn as sn
```

```python
#step-2.import required dataset:

data=pd.read_csv(r"C:\Users\ELCOT\Downloads\diabetes_DS.csv")
data.head()


```

```python
#step-3.preprocessing:

data.info()
```

```python
data.isnull().sum()
```

```python
#step-4.Assing dependent and independent variable:


x=data.iloc[:,0:8]
y=data.iloc[:,-1]

x.shape,y.shape
```

```python
#step-5.Split the data into training and testing:

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
display(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
#step-7.Fit the model:

model=DecisionTreeClassifier(criterion='gini',random_state=5)
model.fit(x_train,y_train)
```

```python
#step-8.Compute and visualize:

y_pred=model.predict(x_test)
print('y_pred:',y_pred)
```

```python
#step-9.Evaluation : confusion matrix:

conf_mat=metrics.confusion_matrix(y_test,y_pred)
print('Confusion Matrix: ',conf_mat)
```

```python
#Accuracy score:

acc_sco=metrics.accuracy_score(y_test,y_pred)
print('Accuracy Score: ',acc_sco)
print('Accuracy in Percentage: ',int(acc_sco*100),'%')
```

```python
#plotting:

conf_mat=pd.crosstab(y_test,y_pred,rownames=['Actual'],colnames=['predicted'])
sn.heatmap(conf_mat,annot=True)
```
